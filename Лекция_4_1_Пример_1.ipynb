# **Приклад 1. Байєсівський класифікатор**

```
import numpy as np
import random
from keras.datasets import mnist
# класифікатор та метрики
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
# дані - зображення рукописних цифр
(X_train, y_train), (X_test, y_test) = mnist.load_data()
#виведемо на екран зображення
X_test_copy=X_test
n_images = X_test.shape[0]
n_rows=4
n_cols=5

for i in range(1,n_rows*n_cols+1):
    im_idx = random.randint(0,n_images-1)
    pixels=X_train[im_idx]
    plt.subplot(n_rows, n_cols, i)
    plt.imshow(pixels, cmap='gray')
    plt.axis('off')
plt.show()
```

# Створимо математичний опис образу. Подамо зображення у векторному вигляді:

```
num_pixels = X_train.shape[1] * X_train.shape[2]
X_train = X_train.reshape(X_train.shape[0], num_pixels).astype('float32')
X_test = X_test.reshape(X_test.shape[0], num_pixels).astype('float32')
```

# Навчимо байєсівський класифікатор та використаємо його

```
clf_nb = MultinomialNB()
# train the model
clf_nb.fit(X_train,y_train)
# make predictions with the NB classifier
y_test_pred_nb = clf_nb.predict(X_test);
a_nb = accuracy_score(y_test, y_test_pred_nb);
print(a_nb)
```
